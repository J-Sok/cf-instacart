## Import libraries & datasets

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# define path variable
path = r'/Users/jsok/Instacart Basket Analysis'
```

```python
# read 'orders_products_merged.pkl' and store in 'ords_prods_merge'
ords_prods_merge = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','orders_products_merged.pkl'))
```

## Bar charts

```python
# create bar chart
# in descending order (by default)
ords_prods_merge['order_day_of_week'].value_counts().plot.bar()
```

```python
# bar chart, sorted by index num
# added color to bars
bar = ords_prods_merge['order_day_of_week'].value_counts().sort_index().plot.bar(color =['red','green','red','green','red','green','red'])
```

```python
# export and save bar chart as a .png file
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

## Scatterplots

```python
# subset of first 2 million rows, 'df'
df = ords_prods_merge[:2000000]
```

```python
# create a scatterplot of prices vs prices
sns.scatterplot(x = 'prices', y = 'prices',data = df)
```

## Histograms

```python
# histogram for 'prices' column
hist = ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# histogram for 'prices' column
# increase granularity
hist_2 = ords_prods_merge['prices'].plot.hist(bins = 70)
```

```python
# export the histogram
hist_2.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','hist_prices.png'))
```

## Sampling data

```python
# set a random seed: 4
np.random.seed(4)
```

```python
# create a list 'dev' that stores boolean values
# 70/30 split
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# list of 32,434,212 boolean values
dev
```

```python
# store 70% of the sample in 'big' df
big = ords_prods_merge[dev]
```

```python
# store 30% of the sample in 'small' df
small = ords_prods_merge[~dev]
```

#### Checking if lengths match

```python
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

## Line chart

```python
# SAMPLE 30% df with only 2 columns necessary for line chart
df_2 = small[['order_day_of_week','prices']]
```

```python
# line chart of 'small' sample df
line = sns.lineplot(data = df_2, x = 'order_day_of_week',y = 'prices')
```

## Export

```python
# export sample line chart
line.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','line_prices_over_dow.png'))
```
